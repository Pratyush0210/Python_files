```python
# input( ) can be used to receive, 1, or more values.
# split( ) function will split the entered fullname with space as a delimiter.

fname, mname, lname = input("Enter your full name: ").split()
print(fname)
print(mname)
print(lname)
```

```python
# more compact manner using a feature called list comprehension.
# n1, n2, n3 = [int(n) for n in input('Enter three values: ').split( )]
# The expression enclosed within [ ] is called list comprehension.
# input( ) can be used to receive different types of values at a time.

fname, mname, lname = [n for n in input("Enter full name: ").split()]
print(fname)
```

```python
n1, n2, n3 = [int(n) for n in input('Enter three values: ').split( )]
print(n1, type(n1))
```

**There are 4 ways to control the formatting of output:**
1.   **Using formatted string literals - easiest**
2.   **Using the format( ) method - older**
3.   **C printf( ) style - legacy**
4.   **Using slicing and concatenation operation - difficult**





```python
# Using formatted string literals - easiest

name = 'chanchal'
place = 'lucknow'
age = 33
salary = 25000.00
print(f"{name} is living in {place}. she is {age} years old. Her salary is Rs.{salary}")
```

```python
name = 'Priya Srivastava'
for n in name.split( ):
  print(f'{n:10}',end='')
```

```python
# Que. print 2, 3, 4, upto 10 table. by using format printing.
```

```python
# Using the format( ) method - older

l, b, h = 15.5678, 10.5675475, 2.66675675
name, age, salary = 'Pratyush', 28, 'berozgar'

# print in order by position of variables using empty {}
print('length = {} breadth = {} height = {}'.format(l, b, h))
print('name = {} age = {} salary = {}'.format(name, age,salary))
```

```python
# print in any desired order
print('breadth = {1} length = {0} height = {2} '.format(l, b, h))
print('age = {1} salary = {2} name = {0} name = {0} Garg'.format(name,age,salary))
```

```python
# print name in 15 columns, salary in 10 columns
print('name = {0:^15} salary = {1:^10}'.format(name, salary))
```

```python
# print length and breadth in 8 columns, with 2 and 5 digits after decimal point
print('length = {0:<8.2f} breadth = {1:<8.5f}'.format(l, b))
```

### **Solved Problems**

```python
# Problem 7.1
# Write a program to receive radius of a circle, and length and breadth of
# a rectangle in one call to input( ). Calculate and print the circumference
# of circle and perimeter of rectangle.

R, L, B = [int(n) for n in input("Enter radius, length and breadth: ").split()]
print(f"Circumference of circle is {2*3.14*R:.2f} and Perimeter of rectangle is {2*(L + B)}")
```

```python
# Problem 7.2
# Write a program to receive 3 integers using one call to input( ). The
# three integers signify starting value, ending value and step value for a
# range. The program should use these values to print the number, its
# square and its cube, all properly right-aligned. Also output the same
# values left-aligned.
```

```python
start, end, step = [int(n) for n in input("Enter Start, end and step values: ").split()]
for n in range(start, end, step):
  print(f"number: {n:<3} square: {n**2:<3} cube: {n**3:<3}")
```

```python
# Problem 7.3
# Write a program to maintain names and phone (cell) numbers of 4 persons and
# then print them systematically in a tabular form.
```

```python
persons = {'chanchal': 9000000000,
           'pratyush': 9415195937,
           'priya': 8000000000,
           'alok': 9876543210}
name, cell_no = 'Name', 'Mobile Number'
print(f"{name:10} : {cell_no:^15}")
for k, v in persons.items():
  print(f"{k:10} : {v:^15}")
```

```python
# Problem 7.4
# Suppose there are 5 variables in a program—max, min, mean, sd and
# var, having some suitable values. Write a program to print these
# variables properly aligned using multiple fstrings, but one call to print( ).

max = 100
min = 1
mean = 50
sd = 3
var = 9

print(f"{'max':5}:{max:^3}\n{'min':5}:{min:^3}\n{'mean':5}:{mean:^3}\n{'sd':5}:{sd:^3}\n{'var':5}:{var:^3}")
```
