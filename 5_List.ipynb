```python
# looping of list

# what is iteration - repeatition

flower = ['Rose', 'Lotus', 'Lilly', 'Jasmine', 'Marrygold', 'Sunflower', 'Tulip']

print(len(flower))
```

```python
# using WHILE loop
i = 0

while i<len(flower):
  print(flower[i])
  i += 1
```

```python
# using FOR loop

for f in flower:
  print(f, end = ' ')
```

```python
# using for loop in complex way

for i in range(len(flower)) :
  print(flower[i], end = ' ')
```

```python
for index, i in enumerate(flower):
  print(index+1, i)
```

# **Built-in Function**

```python
# any()- returns True if any of the elements of a given iterable( List, Dictionary, Tuple, set, etc) are True else it returns False.

lst = [1, 0, False]

print(any(lst))
```

```python
# all()- returns True if all of the elements of a given iterable( List, Dictionary, Tuple, set, etc) are True else it returns False.

lst = [0, 22, 33, 44, 55]

print(all(lst))
```

```python
# del - del()
X = [0, 22, 36, 56, 78, 89]
del X[3]
print(X)
del X[1:3]
print(X)
```
