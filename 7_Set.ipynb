### **Set Methods**

```python
# copy()

s = {0, 5, 10, 15, 20}
u = s.copy()
print(u)
```

```python
# remove()

s0 = {1, 3, 5, 7, 9, 11, 13}

s0.remove(5)
print(s0)
```

```python
# discard() - it would not raise any error if an element is not in the set

s9 = {1, 2, 3, 5, 6, 9}
s9.discard(4)
s9.discard(1)
print(s9)
```

```python
# clear() - remove all elements

s9.clear()
s9
```

```python
# issuperset(), issubset(), isdisjoint()
s = {12, 15, 13, 23, 22, 16, 17}
t = {13, 15, 22}
print(s.issuperset(t))
print(s.issubset(t))
print(t.issubset(s))
print(t.issubset(t))
print(t.isdisjoint(s))
print(s.isdisjoint(set()))
```

### **Mathematical Set Operations**

```python
maths = {"shubh", "kapil", "manas", "dheeraj", "mohini", "pari", "ajay"}
science = {"dheeraj", "pari", "gopal", "anuj", "kapil", "om"}
english = {"shubh", "pari", "kapil", "anuj", "manas", "om"}
```

```python
# union - |
print(maths | science | english)
print(len(maths | science | english))
```

```python
# intersection - &

# maths science
print(maths & science)
```

```python
# maths english
print(maths & english)
```

```python
# science english
print(science & english)
```

```python
# maths science english
print(maths & science & english)
```

```python
# difference - "-"

# maths - science
print(maths - science)
```

```python
# science - maths
print(science - maths)
```

```python
# symmetric difference

print(science ^ maths) # (science - maths) + (maths - science)
```

```python
a = {1, 2, 3, 4, 5}
b = {2, 4, 5}
print(a >= b) # prints True as a is superset of b
print(a <= b) # prints False as a is not a subset of b
```

### **Updating Set Operations**

```python
a = {1, 2, 3, 5}
b = {3, 5, 6, 7}

# a += 1 -> a = a+1

a |= b    # a = a|b
print(a)
```
