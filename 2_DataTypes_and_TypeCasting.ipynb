# **Type Casting**

```python
# str to int 

age = "30"

type(age)
```

```python
age = int(age)
type(age)
```

```python
name = "pratyush"

type(name)
```

```python
name = int(name)
```

```python
# int to float

a = 10
print(type(a))
a = float(a)
```

```python
print(type(a))
print(a)
```

```python
# float to int is not valid 

b = 1.312

type(b)
```

```python
b = int(b)
type(b)
```

```python
print(b)
```

```python
# bool to int, float

a = True
print(a)
```

```python
a = int(True)

print(a)
```

```python
a = float(True)
```

```python
print(a)
```
