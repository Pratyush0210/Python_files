### **Python have 3 ways to perform string formatting:**
- **Placeholders using modulo % character** - Oldest method
- **String.format()** - Improved method
- **f-strings** - Newest Method




### **Formatting with placeholders**
- %s - Strings
- %d - integers
- %f - float values


```python
print("I am writing %s %s"%("Python", "codes"))
```

```python
print('%s %s everywhere'%('data', 'data'))
```

```python
print("I got %1.1f in class %d"%(67.78, 10))
```

```python
# %r - raw string 

print('I am using %s'%'colab\nin google drive')
print('I am using %r'%'colab\nin google drive')
```

### **formatting with string.format()**

```python
print("{} is for {} {}".format('Python', 'Data', 'Science'))
```

```python
print("{2} is for {0} {1} {1} {0} {2} {1}".format('Data', 'Science', 'Python'))
```

```python
print("{p} is for {d} {s} {p}".format( d = 'Data', s = 'Science', p = 'Python'))
```

### **Float formatting follows '{ Value : Width.Precision f }'**

```python
result = 100/77
print("{r} is converted to {r:10.3f}".format(r = result))
```

```python
print("Python is for {0:>10} {1:<20}".format('Data', 'Science'))
```

### **F- String**

```python
name = 'Chanchal'
age = 33
print(f"{name:^20} is {age} years old")
```
