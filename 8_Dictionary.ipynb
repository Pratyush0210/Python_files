### **What are Dictionaries?**


```python
# collection of key-value pairs.
# empty dictionary
s = set()
d1 = {}
d2 = {"01": "Sushma", 
      "02": "Chanchal", 
      "03": "Pratyush", 
      "04": "Priya"}
```

```python
print(d2)
```

```python
# Different keys may have same values.
d = {10 : 'A', 20 : 'A', 30 : 'Z'} 
print(d)
```

```python
# Keys must be unique.
# If keys are same, but values are different, latest key value pair gets stored.

d3= {10 : 'A', 20 : 'B', 10 : 'Z'}
print(d3)
```

```python
d4 = {1: "a", 2: "b", 3: "c", 2: "z"}
print(d4)
```

```python
# If key value pairs are repeated, then only one pair gets stored.
d5 = {10 : 'A', 20 : 'B', 10 : 'A'}
print(d5)
```

### **Looping in Dictionaries**

```python
# items(), keys(), values()

Rollno = {1: "Abhay", 2: "Amit", 3: "Bulbul", 4: "Charit"}
```

```python
a = (1, 4, 67, 89)
x, y, w, t = a
print(x,y,w,t)
```

```python
for k,v in Rollno.items():
  print(k,v)
```

```python
for k in Rollno.keys():
  print(k, end=" ")
```

```python
for k in Rollno:
  print(k, end=" ")
```

```python
for v in Rollno.values():
  print(v, end=" ")
```

```python
# we want index of each key: value pairs - use enumerate() with items()

for i,(k,v) in enumerate(Rollno.items()):
  print(i,k,v)
```

### **Dictionary Operations**

```python
courses = {'CS101' : 'CPP', 
           'CS102' : 'DS', 
           'CS201' : 'OOP',
           'CS226' : 'DAA', 
           'CS601' : 'Crypt', 
           'CS442' : 'Web'}
```

```python
# add new key: value pair.

courses["CS103"] = "Python"
print(courses)
```

```python
# modify any value in given key.

courses["CS201"] = "OOP in Java"
print(courses)
```

```python
# delete any key: value

del courses["CS103"]
```

```python
print(courses)
```

```python
del courses
```

```python
# Nested Dictionary
contacts = {
            'Anil': {'DOB' : '17/11/98', 'Favorite' : 'Igloo'},
            'Amol': {'DOB' : '14/10/99', 'Favorite' : 'Tundra'},
            'Ravi': {'DOB' : '19/11/97', 'Favorite' : 'Artic'}
            }
contacts
```

```python
c = { 'CS101' : 'CPP', 'CS102' : 'DS', 'CS201' : 'OOP'}
d = { 'ME126' : 'HPE', 'ME102' : 'TOM', 'ME234' : 'AEM'}
```

```python
a = {**c, **d}
print(a)
```

```python
b = [*c,*d] # unpack only keys in list form
print(b)
```
