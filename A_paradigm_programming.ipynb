```python
import math as m

def roots(a, b, c):
  d = m.pow(b,2) - 4*a*c
  if d >= 0:
    return (-b + m.sqrt(d))/(2*a), (-b - m.sqrt(d))/(2*a)

  else:
    print("its an imaginary number")
```

```python
a= int(input("a= "))
b= int(input("b= "))
c= int(input("c= "))

roots(a, b, c)
```

```python
roots(10, -5, -6)
```
