## **Basic Tuple Operations**

```python
# Mutability - does not show mutability (Remove(), Append(), insert())

A = (10, 20, 30, 40)

A[2] = 50
```

```python
tpl = ([1, 2, 3, 4], ["dog", "cat", "cow"], [1.5, 2.4, 6.8])
#           0                  1                   2
tpl[0][0] = 10
tpl
```

```python
tpl[1].append('mouse')
tpl
```

```python
tpl[0] = [10,20,30,40]
```

```python
# sorted()

tup = (76, 4, 66, 87, 23, 1, -45)

tup = sorted(tup)

print(tup, type(tup))
```

```python
records = (('Priyanka', 24, 3455.50), 
           ('Shailesh', 25, 4555.50),
           ('Subhash', 25, 4505.50), 
           ('Sugandh', 27, 4455.55))

print(records[0][0], records[0][1], records[0][2])
print(records[1][0], records[1][1], records[1][2])
for n, a, s in records :
  print(n,a,s)
```

```python
# * used to unpack the tuple
x = (1, 2, 3, 4)
y = (10, 20, *x, 30)
print(y) 
```

```python
# sorting in the nested tuple or list

lst = [('Yash', 30, 3455.50), 
       ('Shailesh', 25, 4555.50), 
       ('Chanchal', 30, 6000.50),
       ('Priya', 29, 4500.00)]

l= sorted(lst)
print(l)
```

```python
import operator as op

l1 = sorted(lst, key= op.itemgetter(1), reverse=True)
print(l1)
```
