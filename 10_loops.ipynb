```python
i = 10
while i < 20:
  print(i)
  i += 1
```

```python
i = 1
while i<7:
  print(i, end=" ")
  i += 1
else:
  print("\nthe process is completed")
```

```python
i = 0
while i>= 0:
  if i==10:
    break
  else:
    print(i, end=" ")
    i += 1
else:
  print("process completed")
```

```python
# list 
lst = ["chanchal", "priya", "pratyush", "Rani", "pihu"]
i = 0
while i < len(lst):
  print(lst[i])
  i += 1
else:
  print(".........list completed")
```

```python
count = 0
while count<10:
  print(count, count**2, count**3)
  count += 1
```

```python
num = int(input('Enter a number: '))
while num != 5 :
  print(num, num * num)
  num = int(input('Enter a number: '))
```

### **For loop**

```python
for i in range(20, 10, -3):
  print(i)
```

```python
for i in range(1, 11, 2): # 11 is exclusive
  print(i, i**2, i**3)
```

```python
lst = ["chanchal", "priya", "pratyush", "Rani", "pihu"]
for name in lst:
  print(name)
```

```python
for n in ["chanchal", "priya", "pratyush", "Rani", "pihu"]:
  print(n)
```

```python
for i, name in enumerate(["chanchal", "priya", "pratyush", "Rani", "pihu"]):
  print(i+1, name)
```

```python
for i in range(10):
  print(i, end=" ")
else:
  print("\ncompleted")
```

```python
for i in range(10):
  if i == 7:
    break
  else:
    print(i, end=" ")
else:
  print("\ncompleted")
```

```python
num = int(input('Enter an integer: '))
i = 2
while i <= num - 1 :
  if num % i == 0 :
    print(num, 'is not a prime number')
    break
  i += 1
else :
  print(num, 'is a prime number')
```

```python
for num in range(2,100):
  i = 2
  while i <= num - 1 :
    if num % i == 0 :
#      print(num, 'is not a prime number')
      break
    i += 1
  else :
    print(num,end=" ")
```

```python
number = input("Enter any big digit numbers: ")
rev_num = 0

for n in range(len(number)):
# number, rem = divmod(number, 10)
  number = int(number)
  rem = number % 10
  number = number // 10  # 1234.5 - 1234
  rev_num = rev_num * 10 + rem

print(rev_num)
```

```python
# number = 12345
# rev_num = 0
# 1st iteration - rem = 5, num = 1234, rev_num = 5
# 2nd iteration - rem = 4, num = 123,  rev_num = 54
# 3rd iteration - rem = 3, num = 12,   rev_num = 543
# 4th iteration - rem = 2, num = 1,    rev_num = 5432
# 5th iteration - rem = 1, num = 0,    rev_num = 54321
```
